# What is the maximum & minimum "average_price" per year in england ?


```python
df.head()
```

```python
df['year'] = df['date'].dt.year
england_housing=df[df['area']=='england'].groupby('year')['average_price'].agg(['max','min','mean'])
england_housing
```

```python
england_housing['mean'].plot(kind = "line",color='k',linestyle="dashed",marker = "*")
plt.show()
```
